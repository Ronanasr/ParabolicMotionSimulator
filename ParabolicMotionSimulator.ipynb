```python
import numpy as np
from math import sin, cos, pi
import matplotlib.pyplot as plt

def projectile_coordinates(v0):
    """
    Plot the trajectory of a projectile and the locus of its peak points.

    Parameters:
        v0 (float): Initial velocity of the projectile (m/s).
    """
    # Define gravitational acceleration (m/s^2)
    g = 10

    # Set the launch angle to 45 degrees (pi/4 radians)
    theta = pi / 4
    # Calculate initial velocity components
    vx = v0 * cos(theta)
    vy = v0 * sin(theta)
    # Calculate total flight time
    t_total = 2 * vy / g
    # Generate time array for smooth plotting
    t = np.linspace(0, t_total, 100)
    # Compute x and y coordinates of the parabolic trajectory
    x = vx * t
    y = vy * t - 0.5 * g * t**2

    # Plot the parabolic trajectory in green
    plt.plot(x, y, 'g', label='Parabolic Trajectory')

    # Calculate the locus of peak points for varying angles (forming an ellipse)
    theta_values = np.linspace(0, pi, 100)
    xm_values = (v0**2 / (2 * g)) * np.sin(2 * theta_values)
    ym_values = (v0**2 / (2 * g)) * np.sin(theta_values)**2
    # Plot the locus of peak points as a red line with markers
    plt.plot(xm_values, ym_values, 'ro-', label='Locus of Peak Points')

    # Calculate the peak point of the specific trajectory
    xm = (v0**2 / (2 * g)) * sin(2 * theta)
    ym = (v0**2 / (2 * g)) * sin(theta)**2
    # Plot the peak point as a blue dot
    plt.plot(xm, ym, 'bo', label='Peak of Trajectory')

    # Add labels, title, and legend to the plot
    plt.xlabel('Horizontal Distance (m)')
    plt.ylabel('Vertical Distance (m)')
    plt.title('Projectile Motion and Peak Locus')
    plt.legend()
    plt.grid(True)

    # Save the plot as a PNG file
    plt.savefig('projectile_plot.png', dpi=300, bbox_inches='tight')
    # Display the plot
    plt.show()

# Example usage with initial velocity of 10 m/s
v0 = 10
projectile_coordinates(v0)
```
